```python
# Escreva um programa que recebe 3 valores e escreva os valores em ordem crescente.

# Entrada: Três números inteiros

# Saida: Os três númeors inteiros ordenados de forma crescente.


def ordena_numeros(lista_de_num):

  if lista_de_num[0] > lista_de_num[1]:
    
    if lista_de_num[0] > lista_de_num[2]:
      
      if lista_de_num[1] > lista_de_num[2]:
        # abc
        print(lista_de_num[2],lista_de_num[1],lista_de_num[0])
      else:
        # acb
        print(lista_de_num[1],lista_de_num[2],lista_de_num[0])
    else:
      # cab
        print(lista_de_num[1],lista_de_num[0],lista_de_num[2])
    
  else:
    
    if lista_de_num[0] > lista_de_num[2]:
      #bac
      print(lista_de_num[2],lista_de_num[0],lista_de_num[1])
    else:
      if lista_de_num[1] > lista_de_num[2]:
        # bca
        print(lista_de_num[0],lista_de_num[2],lista_de_num[1])
      else:
        # cba
        print(lista_de_num[0],lista_de_num[1],lista_de_num[2])

n1,n2,n3 = input().split(" ")
ordena_numeros([int(n1),int(n2),int(n3)])
```
