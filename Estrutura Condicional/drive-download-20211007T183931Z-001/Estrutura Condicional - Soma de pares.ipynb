```python
# Desenvolva um algoritmo que leia quatro números inteiros e calcule a soma dos que forem par.

def soma_pares(lista_num):
  
  soma = 0
  
  for n in lista_num:
    
    if n%2 == 0:
      
      soma = soma + n
  
  return(soma)

lista_num = []

for i in range(4):

  lista_num.append(int(input()))

print("A soma dos numeros pares é",soma_pares(lista_num))
```
