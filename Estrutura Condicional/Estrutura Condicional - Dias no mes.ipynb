```python
# Faça um programa que dado o nome do mês, entregue o número de dias que ele tem (calendário 2021).

# Entrada: nome do mês

# Saída: número de dias que tem esse mês no ano de 2021

def dias_do_mes(nome_mes):

  if (nome_mes == "dezembro") or (nome_mes == "outubro") or (nome_mes == "janeiro") or (nome_mes == "março"):
    
    print(31)

  elif (nome_mes == "maio") or (nome_mes == "agosto") or (nome_mes == "julho"):

    print(31)

  elif (nome_mes == "novembro") or (nome_mes == "abril") or (nome_mes == "setembro") or (nome_mes == "junho"):

    print(30)

  else:
    
    print(28)

nome_mes = input()

dias_do_mes(nome_mes)
```
